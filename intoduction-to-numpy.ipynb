### Dot Product

```python
np.random.seed(4)

mat1 = np.random.randint(10, size=(5,3))
mat2 = np.random.randint(10, size=(5,3))

mat1, mat2
```

```python
# Element-wise multipliation (Hadamard product)
mat1 * mat2
```

```python
# Dot product (rows and columns of a matrices must match) so, (5,3) dot (3,5) will work. where as (5,3) dot (5,3) will not work.
# solution should be to Transpose a matrice
mat3 = np.dot(mat1, mat2.T)
mat3, mat3.shape
```

## Dot Product example (nut butter sales)

```python
np.random.seed(0)

# Number of jars sold

sales_amount = np.random.randint(20, size=(5,3))
sales_amount
```

```python
# Create weekly_sales DataFrame

weekly_sales = pd.DataFrame(sales_amount, index=["Mon", "Tues", "Wed", "Thurs", "Fri"],
                               columns=["Almond butter", "Peanut butter", "Cashew butter"])

weekly_sales
```

```python
# Create prices array
prices = np.array([10,8,12])
prices
```

```python
# Create butter_prices DataFrame
butter_prices = pd.DataFrame(prices.reshape(1,3), index=["Price"], columns=["Almond butter", "Peanut butter", "Cashew butter"])

butter_prices
```

```python
# shapes are not aligned, lets transpose
total_sales = prices.dot(sales_amount.T)
total_sales
```

```python
# Create daily_sales
butter_prices.shape, weekly_sales.shape

daily_sales = butter_prices.dot(weekly_sales.T)
daily_sales
```

```python
weekly_sales["Total ($)"] = daily_sales.T
weekly_sales
```
